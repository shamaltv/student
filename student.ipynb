```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, mean_absolute_error, r2_score
import scipy.stats as stats
```

```python
# Load the dataset
csv_path = r"C:\Users\shama\Downloads\A8\archive (10)\Student_performance_data _.csv"
df = pd.read_csv(csv_path)
```

```python
# Exploratory Data Analysis (EDA)
plt.figure(figsize=(10,6))
sns.histplot(df['GPA'], bins=20, kde=True)
plt.title('Distribution of GPA')
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Hypothesis Testing (T-test example: GPA vs Tutoring)
t_stat, p_val = stats.ttest_ind(df[df['Tutoring'] == 1]['GPA'], df[df['Tutoring'] == 0]['GPA'])
print(f'T-test: t-statistic={t_stat}, p-value={p_val}')
```
